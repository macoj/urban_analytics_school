```python
import osmnx as ox
```

## Plotting Exeter residential and tertiary roads

```python
place = "Exeter, UK"
cf = '["highway"~"residential|tertiary"]'
```

```python
G = ox.graph_from_place(place,network_type='drive',custom_filter = cf)
ox.plot_graph(G)
```

## Plotting the NYC railwail data

```python
ox.settings.useful_tags_way += ['railway']
```

```python
G = ox.graph_from_place("New York City, New York",
                        truncate_by_edge = True,
                        simplify = True,
                        custom_filter = '["railway"~"subway"]'
                       )
```

```python
ox.plot_graph(G,node_size=0, edge_color='w', edge_linewidth=0.2)
```
